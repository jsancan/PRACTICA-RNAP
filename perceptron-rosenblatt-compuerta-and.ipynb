Implementar el algoritmo del perceptrón original de Rosenblatt (con función escalón como activación) para aprender la compuerta AND. Mostrar la evolución de los pesos en cada época, verificar la convergencia y calcular la ecuación de la frontera de decisión final.

```python
import numpy as np
```

```python
#Funcion de Activacion Escalon Rosenblatt
def step_function(x): #Se usa para obtener salidas binarias (0 o 1) de forma directa
    return 1 if x >= 0 else 0
#A diferencia de la Sigmoide, esta no es derivable, lo cual es propio del perceptrón original
```

```python
class Perceptron:
    def __init__(self, n_entradas, lr=0.1):
        #Inicializacion en cero
        self.w = np.zeros(n_entradas)
        self.b = 0.0
        self.lr = lr #Learning Rate/Tasa de aprendizaje - Se elige 0.1 para que los ajustes sean graduales y no salten la solución óptima (frontera de decisión

    def predict(self, x):
        z = np.dot(self.w, x) + self.b
        return step_function(z)

    def train(self, X, y, max_epocas=10): #Definimos máximo 10 porque para compuertas lógicas simples,el algoritmo converja (llegue al error 0) en menos de 5 iteraciones.
        print(f"{'Epoca'} | {'Pesos [w1, w2]'} | {'Sesgo (b)'} | {'Errores'}")
        
        for epoca in range(max_epocas):
            errores_en_epoca = 0
            for i in range(len(X)):
                prediccion = self.predict(X[i])
                error = y[i] - prediccion
                
                if error != 0:
                    #Regla de aprendizaje de Rosenblatt: w = w + lr * error * x
                    self.w += self.lr * error * X[i]
                    self.b += self.lr * error
                    errores_en_epoca += 1
            
            print(f"{epoca + 1} | {str(self.w)} | {self.b} | {errores_en_epoca}")
            
            # Verificación de errores
            if errores_en_epoca == 0:
                print(f"\nConvergencia en época {epoca + 1}")
                break
```

```python
# Datos para la compuerta AND
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 0, 0, 1])
```

```python
#Entrenamiento
modelo = Perceptron(n_entradas=2, lr=0.1)
modelo.train(X, y)
```

```python
#Ecuación de la frontera de decision: Se calcula igualando la suma ponderada a cero: w1*x1 + w2*x2 + b = 0
w1, w2 = modelo.w
b = modelo.b
print(f"\nEcuacion final: {w1:.1f}x1 + {w2:.1f}x2 + ({b:.1f}) = 0")
```
